```
import cv2
import os
import numpy as np
import glob
from math import ceil
from matplotlib import pyplot as plt
import matplotlib.image as plt2 
import glob
from scipy.fftpack import fft
```

```
def teste_correlacao(img, filtro, d):
    #lendo a imagem de teste
    half = int((d**2)/2)
    img = cv2.imread(img,0)
    img = cv2.resize(img, (d,d))
    #calculando corrlacao
    img_fft = np.fft.fft2(img)           #transformada da imagem de teste
    img_fft = np.fft.fftshift(img_fft)
    img_fft = img_fft.reshape((d**2,1))  #redimensiona a saida para a dimensao do filtro
    teste = img_fft * np.conj(filtro)    #multiplica a transformada da imagem pelo conjugado do filtro
    teste = np.fft.ifftshift(teste)     
    teste = np.fft.ifft2(teste)          #calcula a inversa   
    #teste = np.fft.ifft2(teste)
    teste = np.real(teste)*d*d           #normaliza os valores
    #x = teste[half]
    teste[half] = teste[0]
    teste[0] = 0
    #print(teste[0])
    return teste
```

```
def obter_mace(img_in, d):
  i = 0
  path = glob.glob('/content/drive/My Drive/Colab Notebooks/Reconhecimento_facial/filtros/*.txt')
  for filtro in path:
    filtro_mace = np.loadtxt(filtro)
    filtro_mace = filtro_mace.reshape(d*d,1)
    B = teste_correlacao(img_in, filtro_mace, d)
    i+=1
    if (B[2048] >= 0.9):
      return print('user '+ str(i) + ' detected')
      #break
    else: continue
  return print('user not detected')
  
    

#filter = np.loadtxt('/content/drive/My Drive/Colab Notebooks/Reconhecimento_facial/filtros/'+)
```
